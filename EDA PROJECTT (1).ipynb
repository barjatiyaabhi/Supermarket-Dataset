## Part-2: 

## Is your supermarket more popular with males or females?

```python
df['Gender'].value_counts() # to check the values of males and females
```

```python
# I used bar chart to check the popularity of the males and females in the supermarket 
x = df['Gender']
y = df['Total']
plt.figure(figsize = (10,7))
plt.bar(x,y,color = 'red')
plt.title('Popularity Of Males & Females',fontsize = 20)
plt.xlabel('Gender',fontsize = 20,color = 'red')
plt.ylabel('Total Purchase',fontsize = 20,color = 'red')
plt.grid(color = 'black', alpha = 0.3, linestyle = '--', linewidth = 2)
plt.show()

# The number of females is more as compare to the males so the populariy of females is higer.
```

## What does the customer rating look like and can you also comment on its skewness? 

```python
df['Rating'].describe() # to describe the Rating column
```

```python
plt.figure(figsize=(10,7))
plt.title('Displot on Customer Rating',fontsize = 20)
sns.distplot(df['Rating'])
plt.xlabel('Rating',fontsize = 20)
plt.ylabel('Density',fontsize = 20)
plt.grid(color = 'black', alpha = 0.3, linestyle = '--', linewidth = 2)
plt.show()

# The rating distribution looks uniform and there seems to be no skewness on the left or right side of the distribution
```

## What can you say about the aggregate sales across the branches? 

```python
plt.figure(figsize=(10,7))
plt.title('Aggregate sales across the Branches',fontsize = 20)
sns.barplot(x = df['Branch'],y = df['cogs'])
plt.xlabel('Branch',fontsize = 20)
plt.ylabel('Cogs',fontsize = 20)
plt.grid(True,color = 'black', alpha = 0.3, linestyle = '--', linewidth = 2)

# Branch C has more number of sales in compare to the other branches
```

## What is the spending pattern of both males and females, meaning in which category do they spend more? 

```python
gender = pd.get_dummies(df['Gender'])
df.head()
# Creating the dummies values for the categorical gender column so that the machine can easiy understand
```

```python
df = pd.concat([df,gender],axis =1)
df.head()
# Merging the new columns i.e Female and Male wih the dataframe by the help of concat function
```

```python
plt.figure(figsize = (13,9))
sns.barplot(x = df['Female'],y = df['Product line'])
plt.title('Females Spending Pattern in Products',fontsize = 20)
plt.xlabel('Females',fontsize = 20)
plt.ylabel('Product Line',fontsize = 20)
plt.grid(True) 

# Females spend more in Fashion accessories
```

```python
plt.figure(figsize = (13,9))
sns.barplot(x = df['Male'],y = df['Product line'])
plt.title('Males Spending Pattern in Products',fontsize = 20)
plt.xlabel('Males',fontsize = 20)
plt.ylabel('Product Line',fontsize = 20)
plt.grid(True) 

# Male spend more in health and beauty
```

## Which day of the week has maximum sales? 


```python
df.Date = pd.to_datetime(df.Date)
# datetime function is used here
```

```python
df['weekday'] = df.Date.dt.weekday
df['Month'] = df.Date.dt.month
df['Year'] = df.Date.dt.year
df['Date'] = df.Date.dt.date

#Seperating Weekday,Month,Year,Date
```

```python
plt.figure(figsize=(8, 6))
plt.title('Daily Sales by Day of the Week')
sns.countplot(df['weekday'])

# Day 05 i.e Saturday has the maximum number of sales because its the weekend.
```

## Which hour of the day is busiest? 

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = (df['Time']).dt.hour
df['Hour'].unique()
```

```python
plt.figure(figsize = (10,7))
sns.lineplot(x="Hour",  y = 'Quantity',data =df).set_title("Product Sales per Hour")
plt.title('Product Sales per Hour',fontsize = 20)
plt.xlabel('Hour',fontsize = 20)
plt.ylabel('Quantity',fontsize = 20)
plt.grid(True,color = 'black', alpha = 0.3, linestyle = '--', linewidth = 2) 

# 14th hour is the busiest hour of the day
```
